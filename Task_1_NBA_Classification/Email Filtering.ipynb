```python
import string
import nltk
import math
import re

import seaborn as sns
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup

import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from collections import Counter, defaultdict
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

nltk.download('stopwords')
```

```python
data=pd.read_csv("emails.csv")
```

```python
data.isnull().sum()
data.head()
```

```python
stopwords_en = stopwords.words("english")
data.drop([col for col in data.columns if col in stopwords_en],axis=1,inplace=True)
data.head()
```

```python
data.shape
```

```python
X = data.iloc[:,1:2868]
X
```

```python
Y = data.iloc[:,-1].values
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB(alpha=1.9)         # alpha by default is 1. alpha must always be > 0. 
# alpha is the '1' in the formula for Laplace Smoothing (P(words))
mnb.fit(X_train,y_train)
y_pred1 = mnb.predict(X_test)
print("Accuracy Score for Naive Bayes : ", accuracy_score(y_pred1,y_test))
```
